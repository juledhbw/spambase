```python
corr=data.corr()
#plt.figure(figsize=(16, 12))
sns.set(rc={'figure.figsize':(16,12)})
ax=sns.heatmap(corr,
            xticklabels=corr.columns,
            yticklabels=corr.columns,
            cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True,
            linewidths=.1)
ax.set(title="all variables correlation heatmap");
```
